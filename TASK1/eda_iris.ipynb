```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df=pd.read_csv("Iris.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
df.duplicated().sum()
```

```python
sns.countplot(data=df, x='Species')
plt.title("Count of Each Iris Species")
plt.show()
```

```python
sns.histplot(df['SepalLengthCm'], kde=True)
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df)
plt.title("Boxplot of Features")
plt.show()
```

```python
sns.pairplot(df, hue='Species')
plt.show()
```

```python
sns.scatterplot(data=df,x='SepalLengthCm',y='SepalWidthCm',hue='Species')
plt.title('Sepal Length v/s Sepal Width')
plt.show()
```

```python
sns.scatterplot(data=df,x='PetalLengthCm',y='PetalWidthCm',hue='Species')
plt.title('Petal Length v/s Petal Width')
plt.show()
```

# Short Report Summary

## 1. CountPlot Summary
### The countplot shows that all three Iris species (Setosa, Versicolor, Virginica) are evenly distributed, with 50 samples each.
### This confirms that the dataset is balanced and suitable for classification without bias.

## 2. Histogram (Histplot) Summary

### The histplots show the distribution of all numerical features:
### a. Sepal_length and sepal_width have moderately spread values with some variation.
### b. Petal_length and petal_width show clearer separation between species, especially Setosa.
### c. No extreme outliers are observed; distributions are smooth and natural.

### These distributions show which features are more informative for separating classes.

## 3. Boxplot Summary

### The boxplots show the range, median, and outliers for each feature:
### a. Setosa consistently has smaller petal measurements than the other two species.
### b. Virginica generally has the largest petal dimensions.
### c. Sepal width varies more compared to other features.
### d. Very few outliers appear, indicating a well-behaved dataset.

### This helps confirm which features differ most between species.

## 4. Pairplot Summary

### The pairplot reveals clear relationships between features:
### a. Setosa forms a completely separate cluster, especially in petal-related features.
### b. Versicolor and Virginica overlap slightly but show visible separation.
### c. Petal_length vs Petal_width offers the strongest separation among species.

### This indicates that petal features are highly predictive for classification.

## 5. Scatterplot Summary

### The scatterplots show relationships between pairs of features, colored by species:
### a. Petal length vs petal width shows strong positive correlation and clear species separation.
### b. Setosa points cluster tightly, while Versicolor and Virginica stretch along a trend line.
### c. The plot visually confirms how well different species can be separated by these features.

### Scatterplots help identify patterns and verify pairplot insights.
